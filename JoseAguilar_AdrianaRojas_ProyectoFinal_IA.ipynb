### ndice Meteorolgico de Incendios Forestales (FWI)

El ndice meteorolgico de incendios forestales (FWI) fue desarrollado por el Servicio Forestal Canadiense, es utilizado por varios pases en el mundo, especialmente en Europa, para clasificar el peligro de incendio e incluye seis componentes que toman en cuenta los efectos de la humedad del combustible y las condiciones climticas en el comportamiento del fuego: [20]

-Cdigo de humedad de combustible fino (FFMC): humedad de la basura y otros combustibles finos curados. Indica la facilidad de encendido y la inflamabilidad del combustible fino.

-Cdigo de humedad de Duff (DMC): humedad promedio de capas orgnicas poco compactadas de profundidad moderada. Indica el consumo de combustible en capas de lona moderadas y material leoso de tamao mediano.

-Cdigo de sequa (DC): humedad promedio de capas orgnicas compactas y profundas. Indica los efectos de la sequa estacional en los combustibles forestales y la cantidad de combustin sin llama en capas de tierra profunda y troncos grandes.

-ndice de propagacin inicial (ISI): tasa esperada de propagacin del fuego. Se basa en la velocidad del viento y FFMC. 

-ndice de acumulacin (BUI): es una clasificacin numrica de la cantidad total de combustible disponible para la combustin. 

-FWI: intensidad del fuego. Se basa en el ISI y el BUI.

### Optimizadores:

El algoritmo ms utilizado para entrenar redes neuronales es el descenso de gradiente estocstico normal, se fija el valor de la tasa de aprendizaje para todas las secuencias recurrentes, por lo que da como resultado una convergencia lenta. Aunque el descenso de gradiente adaptativo ADAGRAD funciona bien para entornos dispersos, se ha observado que su rendimiento se deteriora en entornos donde las funciones de prdida no son convexas y los gradientes son densos debido a la rpida disminucin de la tasa de aprendizaje en estos entornos, ya que utiliza todos los gradientes pasados en la actualizacin. Para abordar este problema, se han propuesto varias variantes de ADAGRAD, como RMSprop, ADAM, ADADELTA, etc., que mitigan la rpida cada de la tasa de aprendizaje utilizando los promedios mviles exponenciales de gradientes pasados cuadrados, limitando esencialmente la dependencia de la actualizacin a solo los ltimos gradientes. [12] 

Para el presente desarrollo, se tomaron en cuenta distintos optimizadores, principalmente el ADAM por ser conocido como el ms robusto, sin embargo finalmente nos decantamos por el RMSprop, por producir mejores resultados y ser recomendado como se investig en [15]. RMSprop es una tcnica de optimizacin basada en gradientes que se utiliza para entrenar redes neuronales. Fue propuesto por el padre de la retropropagacin, Geoffrey Hinton. Los gradientes de funciones muy complejas, como las redes neuronales, tienen una tendencia a desaparecer o explotar a medida que los datos se propagan a travs de la funcin, por lo que RMSprop se desarroll como una tcnica estocstica para el aprendizaje por mini lotes.
RMSprop se ocupa del problema anterior mediante el uso de un promedio mvil de gradientes cuadrados para normalizar el gradiente. Esta normalizacin equilibra el tamao del paso, disminuyendo el paso para gradientes grandes para evitar explosiones y aumentando el paso para gradientes pequeos para evitar desaparecer. Bsicamente, RMSprop utiliza una tasa de aprendizaje adaptativa en lugar de tratar la tasa de aprendizaje como un hiperparmetro por lo que la tasa de aprendizaje va cambiando con el tiempo de forma que este optimizador intenta amortiguar las oscilaciones, pero tambin elimina la necesidad de ajustar la tasa de aprendizaje y lo hace automticamente.
```python
#Se modifica la forma del array y_train
y_train=y_train.values.reshape(y_train.size,1)
```

```python
# Este cdigo se utiliza para la estimacin de la regresin de error caracterstica (REC)
def rec(m,n,tol):
    if type(m)!='numpy.ndarray':
        m=np.array(m)
    if type(n)!='numpy.ndarray':
        n=np.array(n)
    l=m.size
    percent = 0
    for i in range(l):
        if np.abs(10**m[i]-10**n[i])<=tol:
            percent+=1
    return 100*(percent/l)

# Se define la mxima tolerancia permisible para la curva REC en el eje x
tol_max=18
```

# Support Vector Regressor (SVR)

```python
# Se importan bibliotecas para el uso de la mquina de vectores de apoyo.
from sklearn.svm import SVR
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import GridSearchCV
```

```python
scaler = StandardScaler()
```

```python
# Parmetros por evaluar por GridSearchCV
param_grid = {'C': [0.01,0.1,1, 10], 'epsilon': [10,1,0.1,0.01,0.001,0.0001], 'kernel': ['rbf']}
```

```python
# Aplicando la optimizacin grid para los datos suministrados
grid_SVR = GridSearchCV(SVR(),param_grid,refit=True,verbose=0,cv=5)
grid_SVR.fit(scaler.fit_transform(X_train),scaler.fit_transform(y_train))
```

```python
# Se imprimen los mejores parmentros que obtuvo el Grid search
print("Mejores parmetros obtenidos por Grid Search:",grid_SVR.best_params_)
```

```python
# Se entrena el modelo
a=grid_SVR.predict(X_test)
# Se imprime el dato de MSE para la SVR
print("Suma cuadrtica del error para la la mquina de vectores de apoyo:",np.sqrt(np.mean((y_test-a)**2)))
```

```python
# mediante este cdigo se realiza la grfica de la curva de la regresin de error caracterstica (REC) para SVR
rec_SVR=[]
for i in range(tol_max):
    rec_SVR.append(rec(a,y_test,i))

plt.figure(figsize=(5,5))
plt.title("Curva REC para la mquina de vectores de apoyo\n",fontsize=14)
plt.xlabel("Tolerancia absoluta del error para una prediccin ($ha$)")
plt.ylabel("Porcentaje de una prediccin correcta")
plt.xticks([i*5 for i in range(tol_max+1)])
plt.ylim(-10,100)
plt.yticks([i*20 for i in range(6)])
plt.grid(True)
plt.plot(range(tol_max),rec_SVR)
```

# Deep network using Keras & TensorFlow

```python
# Se importan bibliotecas para el uso de la red neuronal densa.
from tensorflow.keras import optimizers
from tensorflow.keras.models import Sequential
import tensorflow.keras.optimizers as opti
from tensorflow.keras.layers import Dense, Activation,Dropout
```

```python
# Se programa la red neuronal, se realizaron distintos experimentos donde se destacan buenos resultados con
# valores de capas ocultas de 100-100-100 neuronas, 100-100-80 neuronas o incluso 64-64-64 neuronas.
# se utilizan valores de dropout entre 0.3 y 0.4
model = Sequential()
model.add(Dense(64, input_dim=12))
model.add(Activation('selu'))
model.add(Dropout(0.3))
model.add(Dense(64))
model.add(Dropout(0.3))
model.add(Activation('selu'))
model.add(Dense(64))
model.add(Activation('elu'))
model.add(Dense(1))
model.summary()
```

```python
# Se define el optimizador a utilizar en este caso el RMSprop y la funcin de prdida que contina siendo MSE
optimizer = opti.RMSprop(learning_rate=0.001)
model.compile(optimizer=optimizer,loss='mse')
```

```python
# Se inicializa el entrenamieto
history = model.fit(X_train, y_train, epochs=100, batch_size=10,verbose=0)
```
