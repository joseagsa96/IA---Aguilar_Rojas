### Índice Meteorológico de Incendios Forestales (FWI)

El índice meteorológico de incendios forestales (FWI) fue desarrollado por el Servicio Forestal Canadiense, es utilizado por varios países en el mundo, especialmente en Europa, para clasificar el peligro de incendio e incluye seis componentes que toman en cuenta los efectos de la humedad del combustible y las condiciones climáticas en el comportamiento del fuego: [20]

-Código de humedad de combustible fino (FFMC): humedad de la basura y otros combustibles finos curados. Indica la facilidad de encendido y la inflamabilidad del combustible fino.

-Código de humedad de Duff (DMC): humedad promedio de capas orgánicas poco compactadas de profundidad moderada. Indica el consumo de combustible en capas de lona moderadas y material leñoso de tamaño mediano.

-Código de sequía (DC): humedad promedio de capas orgánicas compactas y profundas. Indica los efectos de la sequía estacional en los combustibles forestales y la cantidad de combustión sin llama en capas de tierra profunda y troncos grandes.

-Índice de propagación inicial (ISI): tasa esperada de propagación del fuego. Se basa en la velocidad del viento y FFMC. 

-Índice de acumulación (BUI): es una clasificación numérica de la cantidad total de combustible disponible para la combustión. 

-FWI: intensidad del fuego. Se basa en el ISI y el BUI.

### Condiciones temporales y meteorológicas

La ubicación y zonas climáticas específicas son factores que influyen en la capacidad y la velocidad para propagarse un incendio forestal. A medida que las condiciones climáticas globales evolucionan y que se expande el uso de la tierra por parte de los humanos, resulta oportuno analizar las condiciones temporales en el parque Montesinho, ya que las condiciones climáticas mensuales promedio son bastante distintas en el año, además que el día de la semana también podría influir, por ejemplo, días laborales frente a fines de semana, días festivos o feriados, ya que, según estudios, una gran parte de los incendios tienen una causa humana. [23]

Sin embargo, una variedad de factores meteorológicos y ambientales afectan la ocurrencia de incendios forestales. Estos factores incluyen las condiciones climáticas locales, la topografía y el tipo de vegetación, la condición y densidad de esta. Los factores meteorológicos que afectan a los principales incendios forestales son los siguientes: [24]

-Humedad relativa (RH): es la relación entre la presión parcial del vapor de agua y la presión de vapor de equilibrio del agua a una temperatura. Esta afecta directamente sobre el contenido de humedad del combustible. Cuanto mayor sea la humedad relativa, más rápida será la absorción de la humedad del combustible, más lenta la evaporación, aumentará el contenido de humedad del combustible. Por lo que, según el estudio, la humedad relativa se puede utilizar para pronosticar incendios forestales. De forma general, si la HR es mayor que 75% no hay probabilidad de incendio; entre 55% y 75% puede ocurrir un incendio de bajo alcance; de 30% a 55%, puede ser un incendio mayor y una RH menor que 30%, puede ocurrir un incendio catastrófico. En el parque Montesinho, la humedad relativa es alta por la mañana y por la noche; mientras que al mediodía y por la tarde se va al mínimo, por lo que son propensos los incendios forestales en este periodo.

-Precipitación: es el espesor de la lluvia que cae o la nieve que se derrite, afecta directamente el contenido de agua de los materiales combustibles. Si la precipitación es alta y constante a lo largo del año, como las selvas tropicales, hay poca probabilidad de que ocurra un incendio forestal; si está más bien se da de forma marcada con una estación seca larga es propenso a incendios forestales.

-Temperatura: esta afecta directamente los cambios de humedad relativa, si la temperatura aumenta, la presión de vapor saturada de aire aumenta, la humedad relativa disminuye y como ya se mencionó, esto hace más propenso a un incendio. Además, las temperaturas más altas pueden aumentar la temperatura del propio material fósil, por lo que se enciende el material, ocasionando un incendio.

La forma más simple de una red neuronal se conoce como perceptrón, la cual es utilizada principalmente para la clasificación de patrones linealmente separables, es decir, patrones que se encuentran a ambos lados de un hiperplano. Consiste en una neurona artificial con pesos sinápticos y umbral ajustables. El modelo del Perceptrón Multicapa es equivalente a un modelo de regresión lineal, debido a la similitud de la variable de salida que se relaciona aplicando la función de activación sobre una combinación lineal de pesos con las variables de entrada. Cada neurona de la red, excepto la capa de entrada, es un sumatorio de todas sus entradas; que son las salidas de las capas anteriores multiplicadas por unos pesos. A esta suma se le añade un término adicional llamado sesgo o bias. Mediante el entrenamiento, se trata de ajustar estos parámetros aplicando una función no lineal conocida como función de activación [11]. En este desarrollo se considerará perceptrones multicapa con M capas ocultas de N nodos y funciones de activación logística y un nodo de salida con una función lineal [11].

Una máquina de vectores de apoyo (SVM) es un conjunto de algoritmos de aprendizaje supervisado y fueron muy utilizadas antes de la era del aprendizaje profundo. Para muchas aplicaciones se prefería el uso de SVM en lugar de redes neuronales. ### Optimizadores:

El algoritmo más utilizado para entrenar redes neuronales es el descenso de gradiente estocástico normal, se fija el valor de la tasa de aprendizaje para todas las secuencias recurrentes, por lo que da como resultado una convergencia lenta. Existen métodos adaptativos como ADAM y RMSprop, donde la tasa de aprendizaje es variable para cada parámetro. Aunque el descenso de gradiente adaptativo ADAGRAD funciona bien para entornos dispersos, se ha observado que su rendimiento se deteriora en entornos donde las funciones de pérdida no son convexas y los gradientes son densos debido a la rápida disminución de la tasa de aprendizaje en estos entornos, ya que utiliza todos los gradientes pasados ​​en la actualización. Para abordar este problema, se han propuesto varias variantes de ADAGRAD, como RMSprop, ADAM, ADADELTA, etc., que mitigan la rápida caída de la tasa de aprendizaje utilizando los promedios móviles exponenciales de gradientes pasados ​​cuadrados, limitando esencialmente la dependencia de la actualización a solo los últimos gradientes. [12] 

Para el presente desarrollo, se tomaron en cuenta distintos optimizadores, principalmente el ADAM por ser conocido como el más robusto, sin embargo finalmente nos decantamos por el RMSprop, por producir mejores resultados y ser recomendado como se investigó en [15]. RMSprop es una técnica de optimización basada en gradientes que se utiliza para entrenar redes neuronales. Fue propuesto por el padre de la retropropagación, Geoffrey Hinton. Los gradientes de funciones muy complejas, como las redes neuronales, tienen una tendencia a desaparecer o explotar a medida que los datos se propagan a través de la función, por lo que RMSprop se desarrolló como una técnica estocástica para el aprendizaje por mini lotes.
RMSprop se ocupa del problema anterior mediante el uso de un promedio móvil de gradientes cuadrados para normalizar el gradiente. Esta normalización equilibra el tamaño del paso, disminuyendo el paso para gradientes grandes para evitar explosiones y aumentando el paso para gradientes pequeños para evitar desaparecer. Básicamente, RMSprop utiliza una tasa de aprendizaje adaptativa en lugar de tratar la tasa de aprendizaje como un hiperparámetro por lo que la tasa de aprendizaje va cambiando con el tiempo de forma que este optimizador intenta amortiguar las oscilaciones, pero también elimina la necesidad de ajustar la tasa de aprendizaje y lo hace automáticamente. Más aún, elige una tasa de aprendizaje diferente para cada parámetro [16].

### Validación

En aspectos de validación del algoritmo, después de una amplia investigación se definió utilizar la curva de característica de error de regresión (REC) como se recomienda en [9], traza la tolerancia del error en el eje x versus el porcentaje de puntos predichos dentro de la tolerancia en el eje y. El error aquí se define como la diferencia entre el valor predicho y la respuesta del valor real para cualquier punto. Podría ser el cuadrado residual o la desviación absoluta dependiendo de la métrica de error empleada. En nuestro caso, se utiliza la función de pérdida para regresión MSE, pues se adapta de mjor manera a los valores dados.

Caracterizando las curvas REC facilitan la comparación visual de las funciones de regresión entre sí y con el modelo nulo. El área de la curva proporciona una medida válida del rendimiento esperado del modelo de regresión. La curva REC es en gran medida cualitativamente invariante a las opciones de métricas de error y escalado del residual. Escalar la respuesta no cambia el gráfico más que el etiquetado del eje x. Utilizando
Varias métricas de error, como la desviación absoluta o el error al cuadrado, cambian la curva REC, pero la posición relativa de las curvas no cambia cualitativamente. Una función que domina a otra el uso del error al cuadrado también dominará la alternativa que usa la desviación absoluta. 
Estas curvas proporcionan una estrategia eficaz para presentar los resultados a los no expertos, pues en estas, la precisión se define como el porcentaje de puntos que se ajustan a la tolerancia. Si tenemos tolerancia cero, solo aquellos puntos en los que la función se ajusta exactamente se considerarán precisos. Si elegimos una tolerancia que excede el error máximo observado para el modelo en todos los datos, entonces todos los puntos se consideran precisos. Por tanto, existe una clara compensación entre la tolerancia al error y la precisión de la función de regresión. El concepto de tolerancia a errores es atractivo porque la mayoría de los datos de regresión son intrínsecamente inexactos, debido a por ejemplo, errores experimentales y de medición.

```python
# Las variables discretas con más de dos valores no ordenados, como el mes y el día, se transforman
# en una codificación como se recomienda en en paper base de Paulo Cortez Aníbal Morais.
# Codificando los meses
enc = LabelEncoder()
enc.fit(data['month'])
enc.classes_
data['month_encoded']=enc.transform(data['month'])
```

```python
# Definición de entradas y salidas
X_data=data.drop(['area','Log-area','month','day'],axis=1)
y_data=data['Log-area']

print(X_data)
print(y_data)
```

```python
#Se define el conjunto de datos de entrenamiento y prueba. Se probaron test sizes de 0.2, 0.3 0.4 entre otros
X_train, X_test, y_train, y_test = train_test_split(X_data, y_data, test_size=0.3) 
```

```python
#Se modifica la forma del array y_train
y_train=y_train.values.reshape(y_train.size,1)
```

```python
# Este código se utiliza para la estimación de la regresión de error característica (REC)
def rec(m,n,tol):
    if type(m)!='numpy.ndarray':
        m=np.array(m)
    if type(n)!='numpy.ndarray':
        n=np.array(n)
    l=m.size
    percent = 0
    for i in range(l):
        if np.abs(10**m[i]-10**n[i])<=tol:
            percent+=1
    return 100*(percent/l)

# Se define la máxima tolerancia permisible para la curva REC en el eje x
tol_max=18
```

# Support Vector Regressor (SVR)

```python
# Se importan bibliotecas para el uso de la máquina de vectores de apoyo.
from sklearn.svm import SVR
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import GridSearchCV
```

```python
scaler = StandardScaler()
```

```python
# Parámetros por evaluar por GridSearchCV
param_grid = {'C': [0.01,0.1,1, 10], 'epsilon': [10,1,0.1,0.01,0.001,0.0001], 'kernel': ['rbf']}
```

```python
# Aplicando la optimización grid para los datos suministrados
grid_SVR = GridSearchCV(SVR(),param_grid,refit=True,verbose=0,cv=5)
grid_SVR.fit(scaler.fit_transform(X_train),scaler.fit_transform(y_train))
```

```python
# Se imprimen los mejores parámentros que obtuvo el Grid search
print("Mejores parámetros obtenidos por Grid Search:",grid_SVR.best_params_)
```

```python
# Se entrena el modelo
a=grid_SVR.predict(X_test)
# Se imprime el dato de MSE para la SVR
print("Suma cuadrática del error para la la máquina de vectores de apoyo:",np.sqrt(np.mean((y_test-a)**2)))
```

```python
# mediante este código se realiza la gráfica de la curva de la regresión de error característica (REC) para SVR
rec_SVR=[]
for i in range(tol_max):
    rec_SVR.append(rec(a,y_test,i))

plt.figure(figsize=(5,5))
plt.title("Curva REC para la máquina de vectores de apoyo\n",fontsize=14)
plt.xlabel("Tolerancia absoluta del error para una predicción ($ha$)")
plt.ylabel("Porcentaje de una predicción correcta")
plt.xticks([i*5 for i in range(tol_max+1)])
plt.ylim(-10,100)
plt.yticks([i*20 for i in range(6)])
plt.grid(True)
plt.plot(range(tol_max),rec_SVR)
```

# Deep network using Keras & TensorFlow

```python
# Se importan bibliotecas para el uso de la red neuronal densa.
from tensorflow.keras import optimizers
from tensorflow.keras.models import Sequential
import tensorflow.keras.optimizers as opti
from tensorflow.keras.layers import Dense, Activation,Dropout
```

```python
# Se programa la red neuronal, se realizaron distintos experimentos donde se destacan buenos resultados con
# valores de capas ocultas de 100-100-100 neuronas, 100-100-80 neuronas o incluso 64-64-64 neuronas.
# se utilizan valores de dropout entre 0.3 y 0.4
model = Sequential()
model.add(Dense(64, input_dim=12))
model.add(Activation('selu'))
model.add(Dropout(0.3))
model.add(Dense(64))
model.add(Dropout(0.3))
model.add(Activation('selu'))
model.add(Dense(64))
model.add(Activation('elu'))
model.add(Dense(1))
model.summary()
```

```python
# Se define el optimizador a utilizar en este caso el RMSprop y la función de pérdida que continúa siendo MSE
optimizer = opti.RMSprop(learning_rate=0.001)
model.compile(optimizer=optimizer,loss='mse')
```

```python
# Se inicializa el entrenamieto
history = model.fit(X_train, y_train, epochs=100, batch_size=10,verbose=0)
```
