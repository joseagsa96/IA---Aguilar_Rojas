## Introduccin
Una red neuronal convolucional es de los algoritmos ms usados en la identificacin y anlisis de imgenes por computadora, o mejor conocido como visin por computadores. Esta arquitectura, est compuesta por varias capas, las cuales cada una posee una funcin especfica para la clasificacin de imgenes. Estas redes, empiezan primeramente con una imagen de entrada, a la cual se le aplica el proceso matemtico de convolucin, y luego una funcin de activacin no lineal, seguidamente posee la capa de pooling, y finalmente, se tiene una red neuronal densa, como las previamente vistas en clases (MLP). [1]
La operacin de convolucin consiste en, por medio del uso de un filtro de convolucin o kernel [2], obtener caractersticas de la imagen. Este proceso de convolucin es el producto punto  en distintas partes de la imagen, y da como resultado un conjunto de mapas de activacin. La operacin de pooling, es realizar un downsampling de la imagen, por medio del submuestreo  de los mapas de activacin previamente mencionados, y as, se tienen las caractersticas dadas por la capa convolucional, pero de una forma ms generalizada [2]. Finalmente, se procede a la fase en la que se tiene una red neuronal densa, y esta permite clasificar los datos que se obtuvieron por medio de la red neuronal convolucional, y tiene como salida un vector con probabilidades, las cuales representa que tan probable es que la imagen analizada pertenezca a cierta categora.

Seguidamente se crea el modelo en s de la red CNN, con una capa convolucional, una capa de pooling, una capa densa y otra capa densa ms de salida. La operacin de convolucin consiste en obtener caractersticas de la imagen mediante el uso de filtros de convolucin o kernel (8, 16 o 24 filtros, en nuestro caso) de un tamao determinado. Este proceso de convolucin es el producto punto en distintas partes de la imagen, y da como resultado un conjunto de mapas de activacin, estos mapas se utilizarn para el anlisis ms adelante. La operacin de pooling realiza un submuestreo de los mapas de activacin para obtener caractersticas dadas por la capa convolucional de una forma ms generalizada, se puede dar en maxpooling, minpooling, o average pooling. Una capa densa clasifica los datos que se obtuvieron por medio de la CNN y se obtiene como salida un vector con probabilidades que representa que tan probable es que la imagen analizada pertenezca a cierta categora. Para el caso de la red densa, adems, se tienen como variables el nmero de neuronas por capa neuronal y la cantidad de capas, parmetros que se comentaran ms adelante. Con estos modelos se determinan los valores de exactitud y perdida obtenidos para cada red, y as, mediante anlisis de estas, concluir cual red da mejores resultados para la clasificacin.
### Determinacin de los factores de influencia
#### Tasa de aprendizaje:
Este parmetro determina que tan rpido se entrena la red densa. La tasa de aprendizaje afecta los saltos que da el algoritmo de descenso de gradiente, para as encontrar un mnimo global a la funcin de error. Si se elige un valor muy grande, la red aprende demasiado rpido y el error podra aumentar en vez de disminuir conforme esta se entrena.
#### Nmero de iteraciones:
Determina la cantidad de veces que se entrena la red neuronal, hasta lograr encontrar un valor de error que sea satisfactorio. Este nmero depender del problema a resolver y del valor de otros factores de influencia como lo es la tasa de aprendizaje, ya que, si se selecciona una muy pequea, se deber iterar ms para entrenar la red de forma ptima.
#### Nmero de neuronas en la capa densa:
Para la capa densa y la red densa, se define la cantidad de neuronas en estas capas, para as lograr obtener un resultado de la clasificacin ms satisfactorio. El nmero deber ser determinado de manera que se obtengan los mejores resultados posibles para el problema a resolver 
#### Nmero de filtros o kernel:
El nmero de filtros en una red convolucional es equivalente a agregar neuronas en la capa oculta de una red neuronal densa tipo perceptrn multicapa. Cada kernel, extrae una caracterstica de la imagen, por lo que entre ms kernels se tengan, la red a detectar ms caractersticas, por lo que la cantidad de filtros depender del problema a resolver
#### Tamao del filtro o kernel:
Determina la regin en la imagen de entrada sobre la cual trabaja el kernel o el campo receptivo de la red. Dependiendo del tamao del kernel, cambian las dimensiones de la salida en la capa convolucional, por consecuencia, el tamao de la entrada en la siguiente capa de la red. Si se usa un kernel muy grande se usa pocos recursos computacionales, pero, se pierden detalles de la imagen, adems, el error es propenso a aumentar y disminuye la exactitud. Si se usa un filtro de tamao muy pequeo, se tienen muchos detalles en la imagen, sin embargo, la red puede ser propensa al sobre entrenamiento.
#### Nmero de capas convolucionales:
En una CNN, las capas extraen caractersticas de la imagen que entra a esa capa por medio de los filtros o kernels. Es comn que la primera capa realice una labor simple de extraccin, identificando caractersticas como bordes, curvas, entre otros. Aumentando el nmero de capas, se realizan operaciones ms complejas, debido a que se est operando sobre ya una imagen que fue modificada por la capa anterior.

```python
#se carga dataset de mnist

from emnist import list_datasets

from emnist import extract_test_samples
from emnist import extract_training_samples

#Se usa el conjunto de datos letters

training_images, training_labels = extract_training_samples('letters')
test_images, test_labels = extract_test_samples('letters')
```

```python
#Muestra la informacion del set de entrenamiento
print('Training Dataset Shape: {}'.format(training_images.shape))
print('No. of Training Dataset Labels: {}'.format(len(training_labels)))
print('Test Dataset Shape: {}'.format(test_images.shape))
print('No. of Test Dataset Labels: {}'.format(len(test_labels)))

training_images = training_images / 255.0
test_images = test_images / 255.0
print((training_labels.shape))
print(np.max(training_labels))
print(np.min(training_labels))
```

```python
# Funcion para reducir el set de letters a 2600 de entrenamiento y 1560 de prueba. Esto debido
# a que son 26 letras con 100 ejemplos de cada letra para entrenar y 60 ejemplares de cada letra
# para validar. Cada uno de estos son tomados al azar.

def AzarLetras():
    indexTrain = np.zeros((1,26, 100))          #lista de ceros para almacenar indices al azar de las 26 letras
    indexTest = np.zeros((1,26, 60))            #100 y 60 ejemplo, para train y test, respectavimante

  #seleccion de 100 ejemplos por digito, al azar

  #Ajuste para set de training
    for x in range (np.min(training_labels),np.max(training_labels)+1): #Itera por cada letra del 1 al 26
        indexes = np.where(training_labels == x)[0]                     #Donde la etiqueta de una letra sea igual al "x"
                                                                    #que se esta analizando, guarda el indice de este
        randInd = random.sample(range(indexes.shape[0]), 100)           #Agarra unicamente 100 ejemplares de la lista 
                                                                    #anterior al azar
        indexes = indexes[randInd]                                      #Utiliza unicamente los indices seleccionados.
        indexTrain[0][x-1] = indexes                                      #Agrega los indices a usar para una de las letras,
                                                                    #que van de la A a la Z, en la lista de indices de
                                                                    #todas las letras

    print(indexTrain.shape)                                           #Imprime la forma que tiene la lista donde se 
                                                                    #guardan los ndices del conjunto de entrenamiento
    indexTrain = indexTrain.reshape((2600,))                          #Se ajusta la forma de la lista del conjunto de 
                                                                    #entrenamiento para que queden todos los indices 
                                                                    #seguidos en "una dimensin"
    indexTrain = indexTrain.astype(int)                               #Se ajusta el tipo detos de la lista
  
    print(indexTrain.shape) 
  
  #Ajuste para set de prueba                                    #Se realiza el mismo procedimiento que en el set de training
    for i in range (np.min(test_labels),np.max(test_labels)+1):
        indexes = np.where(test_labels == i)[0]
        randInd = random.sample(range(indexes.shape[0]), 60)
        indexes = indexes[randInd]
        indexTest[0][i-1] = indexes

    print(indexTest.shape) 

    indexTest = indexTest.reshape((1560,))
    indexTest = indexTest.astype(int)

    print(indexTest.shape)

    return indexTrain, indexTest
```

```python
indexTrain, indexTest = AzarLetras()

#se reduce el training set a los ejemplos seleccionados
training_images = training_images[indexTrain,:]
training_labels = training_labels[indexTrain]

#se reduce el test set a los ejemplos seleccionados
test_labels = test_labels[indexTest]
test_images = test_images[indexTest,:]


#shuffle a los set de imagenes
#ayuda porque validation_split no funciona adecuadamente si estan ordenados
training_images, training_labels = shuffle(training_images, training_labels)
test_images, test_labels = shuffle(test_images, test_labels)
```

```python
#creacion del modelo

numFiltros = 24                   #nucleos convolucion kernels
model = ks.models.Sequential()

#Configuracion de la capa de convolucin
model.add(ks.layers.Conv2D(numFiltros, (7, 7), activation='relu', input_shape=(28, 28, 1), name='Conv2D_layer1'))

#model.add(ks.layers.Conv2D(numFiltros, (2, 2), activation='relu', input_shape=(28, 28, 1), name='Conv2D_layer2'))


#Configuracin de la capa de pooling
model.add(ks.layers.MaxPooling2D((4, 4), name='Maxpooling_2D'))
model.add(ks.layers.Flatten(name='Flatten'))                                    #Se pasan los datos a "una sola dimensin"
                                                                                 #para que cada pixel sea una entrada de la 
                                                                                 #parte densa de una red convolucional
model.add(ks.layers.Dense(100, activation='relu', name='Hidden_layer1'))          #50 numero de neuronas en capa oculta


#Configuracin de la capa de salida
model.add(ks.layers.Dense((np.max(training_labels)+1), activation='softmax', name='Output_layer'))


##modelo de red densa
#model = ks.Sequential()                                    
#model.add(ks.Input(shape=(28, 28, 1))) #(784)                        #Capa de entrada de la red densa, las imagenes son
                                                                     #de 28x28 pixeles, para un total de 784 pixeles,
                                                                     #donde cada uno es una entrada
#model.add(ks.layers.Flatten(name='Flatten'))                         
#model.add(ks.layers.Dense(64, activation="relu", name='Hidden_layer'))    # Configuracin de la capa densa con 70 neuronas
                                                                          #tf.nn.relu, name='Hidden_layer'))
#model.add(ks.layers.Dense(27, activation="softmax", name='Output_layer')) #Configuracion de la capa de salida con 26 neurona,
                                                                          #correspondiente a cada letra
                                                                          #tf.nn.softmax, name='Output_layer'))

model.summary()
```

```python
#se compila y entrena el modelo
inicio = time()
model.compile(loss='sparse_categorical_crossentropy', optimizer = tf.keras.optimizers.Adam(learning_rate=0.001), metrics=['accuracy'])

#Para redes convolucionales
history = model.fit(training_images, training_labels, epochs=50, validation_split=0.01, batch_size=100)

##Para red densa
#history = model.fit(training_images, training_labels, epochs=25, verbose=1, validation_data = (test_images, test_labels))


final=time()

print("Tiempo de ejecucin:", final-inicio)
```
