## Introducción
Una red neuronal convolucional es de los algoritmos más usados en la identificación y análisis de imágenes por computadora, o mejor conocido como visión por computadores. Esta arquitectura, está compuesta por varias capas, las cuales cada una posee una función específica para la clasificación de imágenes. Estas redes, empiezan primeramente con una imagen de entrada, a la cual se le aplica el proceso matemático de convolución, y luego una función de activación no lineal, seguidamente posee la capa de pooling, y finalmente, se tiene una red neuronal densa, como las previamente vistas en clases (MLP). [1]
La operación de convolución consiste en, por medio del uso de un filtro de convolución o kernel [2], obtener características de la imagen. Este proceso de convolución es el producto punto ´ en distintas partes de la imagen, y da como resultado un conjunto de mapas de activación. La operación de pooling, es realizar un “downsampling” de la imagen, por medio del submuestreo ´ de los mapas de activación previamente mencionados, y así, se tienen las características dadas por la capa convolucional, pero de una forma más generalizada [2]. Finalmente, se procede a la fase en la que se tiene una red neuronal densa, y esta permite clasificar los datos que se obtuvieron por medio de la red neuronal convolucional, y tiene como salida un vector con probabilidades, las cuales representa que tan probable es que la imagen analizada pertenezca a cierta categoría.

Seguidamente se crea el modelo en sí de la red CNN, con una capa convolucional, una capa de pooling, una capa densa y otra capa densa más de salida. La operación de convolución consiste en obtener características de la imagen mediante el uso de filtros de convolución o kernel (8, 16 o 24 filtros, en nuestro caso) de un tamaño determinado. Este proceso de convolución es el producto punto en distintas partes de la imagen, y da como resultado un conjunto de mapas de activación, estos mapas se utilizarán para el análisis más adelante. La operación de pooling realiza un submuestreo de los mapas de activación para obtener características dadas por la capa convolucional de una forma más generalizada, se puede dar en maxpooling, minpooling, o average pooling. Una capa densa clasifica los datos que se obtuvieron por medio de la CNN y se obtiene como salida un vector con probabilidades que representa que tan probable es que la imagen analizada pertenezca a cierta categoría. Para el caso de la red densa, además, se tienen como variables el número de neuronas por capa neuronal y la cantidad de capas, parámetros que se comentaran más adelante. Con estos modelos se determinan los valores de exactitud y perdida obtenidos para cada red, y así, mediante análisis de estas, concluir cual red da mejores resultados para la clasificación.
### Determinación de los factores de influencia
#### Tasa de aprendizaje:
Este parámetro determina que tan rápido se entrena la red densa. La tasa de aprendizaje afecta los saltos que da el algoritmo de descenso de gradiente, para así encontrar un mínimo global a la función de error. Si se elige un valor muy grande, la red aprende demasiado rápido y el error podría aumentar en vez de disminuir conforme esta se entrena.
#### Número de iteraciones:
Determina la cantidad de veces que se entrena la red neuronal, hasta lograr encontrar un valor de error que sea satisfactorio. Este número dependerá del problema a resolver y del valor de otros factores de influencia como lo es la tasa de aprendizaje, ya que, si se selecciona una muy pequeña, se deberá iterar más para entrenar la red de forma óptima.
#### Número de neuronas en la capa densa:
Para la capa densa y la red densa, se define la cantidad de neuronas en estas capas, para así lograr obtener un resultado de la clasificación más satisfactorio. El número deberá ser determinado de manera que se obtengan los mejores resultados posibles para el problema a resolver 
#### Número de filtros o kernel:
El número de filtros en una red convolucional es equivalente a agregar neuronas en la capa oculta de una red neuronal densa tipo perceptrón multicapa. Cada kernel, extrae una característica de la imagen, por lo que entre más kernels se tengan, la red a detectará más características, por lo que la cantidad de filtros dependerá del problema a resolver
#### Tamaño del filtro o kernel:
Determina la región en la imagen de entrada sobre la cual trabaja el kernel o el campo receptivo de la red. Dependiendo del tamaño del kernel, cambian las dimensiones de la salida en la capa convolucional, por consecuencia, el tamaño de la entrada en la siguiente capa de la red. Si se usa un kernel muy grande se usa pocos recursos computacionales, pero, se pierden detalles de la imagen, además, el error es propenso a aumentar y disminuye la exactitud. Si se usa un filtro de tamaño muy pequeño, se tienen muchos detalles en la imagen, sin embargo, la red puede ser propensa al sobre entrenamiento.
#### Número de capas convolucionales:
En una CNN, las capas extraen características de la imagen que entra a esa capa por medio de los filtros o kernels. Es común que la primera capa realice una labor simple de extracción, identificando características como bordes, curvas, entre otros. Aumentando el número de capas, se realizan operaciones más complejas, debido a que se está operando sobre ya una imagen que fue modificada por la capa anterior.

```python
#se carga dataset de mnist

from emnist import list_datasets

from emnist import extract_test_samples
from emnist import extract_training_samples

#Se usa el conjunto de datos letters

training_images, training_labels = extract_training_samples('letters')
test_images, test_labels = extract_test_samples('letters')
```

```python
#Muestra la informacion del set de entrenamiento
print('Training Dataset Shape: {}'.format(training_images.shape))
print('No. of Training Dataset Labels: {}'.format(len(training_labels)))
print('Test Dataset Shape: {}'.format(test_images.shape))
print('No. of Test Dataset Labels: {}'.format(len(test_labels)))

training_images = training_images / 255.0
test_images = test_images / 255.0
print((training_labels.shape))
print(np.max(training_labels))
print(np.min(training_labels))
```

```python
# Funcion para reducir el set de letters a 2600 de entrenamiento y 1560 de prueba. Esto debido
# a que son 26 letras con 100 ejemplos de cada letra para entrenar y 60 ejemplares de cada letra
# para validar. Cada uno de estos son tomados al azar.

def AzarLetras():
    indexTrain = np.zeros((1,26, 100))          #lista de ceros para almacenar indices al azar de las 26 letras
    indexTest = np.zeros((1,26, 60))            #100 y 60 ejemplo, para train y test, respectavimante

  #seleccion de 100 ejemplos por digito, al azar

  #Ajuste para set de training
    for x in range (np.min(training_labels),np.max(training_labels)+1): #Itera por cada letra del 1 al 26
        indexes = np.where(training_labels == x)[0]                     #Donde la etiqueta de una letra sea igual al "x"
                                                                    #que se esta analizando, guarda el indice de este
        randInd = random.sample(range(indexes.shape[0]), 100)           #Agarra unicamente 100 ejemplares de la lista 
                                                                    #anterior al azar
        indexes = indexes[randInd]                                      #Utiliza unicamente los indices seleccionados.
        indexTrain[0][x-1] = indexes                                      #Agrega los indices a usar para una de las letras,
                                                                    #que van de la A a la Z, en la lista de indices de
                                                                    #todas las letras

    print(indexTrain.shape)                                           #Imprime la forma que tiene la lista donde se 
                                                                    #guardan los índices del conjunto de entrenamiento
    indexTrain = indexTrain.reshape((2600,))                          #Se ajusta la forma de la lista del conjunto de 
                                                                    #entrenamiento para que queden todos los indices 
                                                                    #seguidos en "una dimensión"
    indexTrain = indexTrain.astype(int)                               #Se ajusta el tipo detos de la lista
  
    print(indexTrain.shape) 
  
  #Ajuste para set de prueba                                    #Se realiza el mismo procedimiento que en el set de training
    for i in range (np.min(test_labels),np.max(test_labels)+1):
        indexes = np.where(test_labels == i)[0]
        randInd = random.sample(range(indexes.shape[0]), 60)
        indexes = indexes[randInd]
        indexTest[0][i-1] = indexes

    print(indexTest.shape) 

    indexTest = indexTest.reshape((1560,))
    indexTest = indexTest.astype(int)

    print(indexTest.shape)

    return indexTrain, indexTest
```

```python
indexTrain, indexTest = AzarLetras()

#se reduce el training set a los ejemplos seleccionados
training_images = training_images[indexTrain,:]
training_labels = training_labels[indexTrain]

#se reduce el test set a los ejemplos seleccionados
test_labels = test_labels[indexTest]
test_images = test_images[indexTest,:]


#shuffle a los set de imagenes
#ayuda porque validation_split no funciona adecuadamente si estan ordenados
training_images, training_labels = shuffle(training_images, training_labels)
test_images, test_labels = shuffle(test_images, test_labels)
```

```python
#creacion del modelo

numFiltros = 24                   #nucleos convolucion kernels
model = ks.models.Sequential()

#Configuracion de la capa de convolución
model.add(ks.layers.Conv2D(numFiltros, (7, 7), activation='relu', input_shape=(28, 28, 1), name='Conv2D_layer1'))

#model.add(ks.layers.Conv2D(numFiltros, (2, 2), activation='relu', input_shape=(28, 28, 1), name='Conv2D_layer2'))


#Configuración de la capa de pooling
model.add(ks.layers.MaxPooling2D((4, 4), name='Maxpooling_2D'))
model.add(ks.layers.Flatten(name='Flatten'))                                    #Se pasan los datos a "una sola dimensión"
                                                                                 #para que cada pixel sea una entrada de la 
                                                                                 #parte densa de una red convolucional
model.add(ks.layers.Dense(100, activation='relu', name='Hidden_layer1'))          #50 numero de neuronas en capa oculta


#Configuración de la capa de salida
model.add(ks.layers.Dense((np.max(training_labels)+1), activation='softmax', name='Output_layer'))


##modelo de red densa
#model = ks.Sequential()                                    
#model.add(ks.Input(shape=(28, 28, 1))) #(784)                        #Capa de entrada de la red densa, las imagenes son
                                                                     #de 28x28 pixeles, para un total de 784 pixeles,
                                                                     #donde cada uno es una entrada
#model.add(ks.layers.Flatten(name='Flatten'))                         
#model.add(ks.layers.Dense(64, activation="relu", name='Hidden_layer'))    # Configuración de la capa densa con 70 neuronas
                                                                          #tf.nn.relu, name='Hidden_layer'))
#model.add(ks.layers.Dense(27, activation="softmax", name='Output_layer')) #Configuracion de la capa de salida con 26 neurona,
                                                                          #correspondiente a cada letra
                                                                          #tf.nn.softmax, name='Output_layer'))

model.summary()
```

```python
#se compila y entrena el modelo
inicio = time()
model.compile(loss='sparse_categorical_crossentropy', optimizer = tf.keras.optimizers.Adam(learning_rate=0.001), metrics=['accuracy'])

#Para redes convolucionales
history = model.fit(training_images, training_labels, epochs=50, validation_split=0.01, batch_size=100)

##Para red densa
#history = model.fit(training_images, training_labels, epochs=25, verbose=1, validation_data = (test_images, test_labels))


final=time()

print("Tiempo de ejecución:", final-inicio)
```
