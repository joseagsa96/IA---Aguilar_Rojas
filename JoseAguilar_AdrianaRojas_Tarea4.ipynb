## INTRODUCCIÓN

La computación evolutiva tiene como objetivo imitar procesos de la evolución natural, donde el concepto principal es la supervivencia del más apto: el débil debe morir. En la evolución natural, la supervivencia se logra mediante la reproducción. La descendencia, reproducida de dos o más padres, contiene material genético de estos. Aquellos individuos que heredan malas características son débiles y pierden la batalla por sobrevivir. [3]

Para cada generación, los individuos compiten por reproducir descendencia. Los individuos con mejor capacidad de supervivencia tienen más posibilidad de reproducirse. La descendencia se genera mediante la combinación de partes de los padres, un proceso denominado cruce. [3]

Después de cada generación, los individuos pueden ser sacrificados o los individuos pueden sobrevivir hasta la siguiente generación (lo que se conoce como elitismo). Además, las características del comportamiento se pueden utilizar para influir en el proceso evolutivo.[3]

Los algoritmos genéticos permiten lograr encontrar soluciones a problemas en los cuales es complejo llegar a un resultado mediante métodos analíticos. Estos algoritmos son procesos de optimización, al igual que la evolución, en la cual la meta es mejorar la habilidad de supervivencia de un individuo.
Los algoritmos evolutivos utilizan una población de individuos, denominados cromosomas los cuales definen sus características que se conocen como gen. El valor de un gen se denomina alelo. Estas características se refieren a las variables del problema de optimización, para las que se busca una asignación óptima. [3]

Los algoritmos evolutivos tienen parámetros que afectan el resultado obtenido, por lo que se debe llevar a cabo un estudio del efecto de cada uno de estos para encontrar la mejor estructura. Algunos de estos parámetros son la población inicial, el método de recombinación, el tipo de selección para los padres, y el tipo y el grado de mutación. Estas variables afectan la presión selectiva y la variabilidad del algoritmo. La presión selectiva se refiere a qué tanto se tarda en producir una población uniforme o dicho de otra forma, converger a una solución y la variabilidad significa qué tantas soluciones especialmente variadas posee el algoritmo. ### Relación genotipo – fenotipo: tratamiento de ambos conceptos en este problema en particular. 
Las carácterísticas de un individuo se pueden dividir en dos clases de informacion evolutiva: genotipos y fenotipos. Un genotipo describe la composición genética de un individuo heredada de sus padres y representa qué alelo posee el individuo, en este caso específico de valores de interés de los parámetros (Ax, Bx, Cx, Dx, Kx y Fx).
Por otra parte, un fenotipo son los rasgos de comportamiento expresados de un individuo en un entorno específico y define cómo se ve un individuo. 
En este caso, las variables mencionadas componen el modelaje de un sistema de control en el dominio del tiempo. Se sabe que el cromosoma se define por Ax, Bx, Cx, Dx, Kx y Fx, por o que el genotipo en este caso son las distintas configuraciones del cromosoma, es decir, que el genotipo son los valores de Ax, Bx, Cx, Dx, Kx y Fx. Por su parte, el fenotipo se refiere a cómo se comportará el sistema de control en el dominio del tiempo.

```python
#Para definir el modelo y la funcion de transferencia
#Se crea una clase porque cada nueva funcion de transferencia que se forme
#tiene características
class funct_transf:
    def __init__(self):
        self.sys = None #funcion transf sistema
        self.estable = None #estabilidad (bool)
        self.polos = None #polos obtenidos
        self.polos_real = None #parte real polos
        self.polos_modif = None
        self.sobreimpulso = None
        self.p_time = None
        self.mejor_sol = None
        self.last_fitness = 0
ft = funct_transf()
```

```python
#Para revisar estabilidad de polos (la parte real de los polos debe estar a la izquierda del eje)
def es_estable(polos_real):
    for polo_real in polos_real:
        if polo_real >= 0:
            return False 
    return True
```

```python
#Para verificar que el sistema que se obtiene sea estable 
def sacar_modelo(solution): #solution es el vector gen solución de cada generación
    A1=solution[0]
    A2=solution[1]
    A3=solution[2]
    B1=solution[3]
    B2=solution[4]
    B3=solution[5]
    C1=solution[6]
    C2=solution[7]
    C3=solution[8]
    D2=solution[9]
    K1=solution[10]
    K2=solution[11]
    K3=solution[12]
    F1=solution[13]
    F2=solution[14]
    s = ctl.tf('s') #crea el operador diferencial s
    G1= A1/(B1*s*(s+C1))
    G2= A2*s/(B2*s*(s+(C2/s))+D2)
    G3= (A3*s*(s+C3))/(B3)
    H1= K1*(1+F1*s)
    H2= K2*s*(1+F2*s)
    H3= K3*s
    #funcion transferencia todo el sistema
    STF= ctl.minreal((G1*G2*G3+G1*G3*H1)/
                    ((1+G1*G2*H2+G1*G2*G3*H3+G1*G3*H1*H3+G1*G2*G3+G1*G3*H1)
                    +0.00000001),verbose=False)
#ctl.minreal(STF) para eliminar estados incontrolables o
# inobservables y cancelar pares de polos y ceros
    return STF
```

```python
#Definiendo la función de calidad
def fitness_func(solution, solution_idx):
    sys = sacar_modelo(solution) #con la solucion actual se crea el modelo
    ft.polos = ctl.pole(sys) #sacar polos
    ft.polos_real = np.real(ft.polos)#parte real de polos para probar si estable
    ft.estable = es_estable(ft.polos_real)
    if not ft.estable: 
        """si no es estable no se puede usar la misma funct calidad por que no hay 
         #tiempo de estabilizacion"""
        ft.polos_modif = np.copy(ft.polos_real)
        ft.polos_modif = ft.polos_modif[ft.polos_modif >= 0] ### POLOS POSITIVOS
        fitness = -1000000*np.sum(ft.polos_modif) ##CASTIGO
    else:
    #Caso estable
    #print("--------------------------------------------------------Estable")
        resp_escalon = ctl.step_info(sys) #Respuesta al escalón
        ft.sobreimpulso = resp_escalon["Overshoot"] #Sobreimpulso obtenido
        ft.p_time = resp_escalon["PeakTime"] #Tiempo donde se da sobreimpulso
        if (ft.sobreimpulso==0):
            output = 2000000 #Caso de castigo
            fitness = 1 / output 
        else:
            #FUNCIÓN CALIDAD
            output = np.absolute(20-ft.sobreimpulso)/4 + np.absolute(1-ft.p_time)*3
            fitness = 100/np.absolute(output)
    return fitness
```

```python
#Definiendo lo que se va a ir informando por generación en la instancia
def callback_generation(ga_instance):
    print("Generation = {generation}".format(generation=ga_instance.generations_completed))
    print("Fitness = {fitness}".format(fitness=ga_instance.best_solution()[1]))
    print("Change = {change}".format(change=ga_instance.best_solution()[1] - ft.last_fitness))
    ft.last_fitness = ga_instance.best_solution()[1]
    #Obtener la mejor solución de cada generación e imprimirla
    ft.mejor_sol = list(ga_instance.best_solution()[0])
    ft.sys = sacar_modelo(ft.mejor_sol)
    ft.polos = ctl.pole(ft.sys)
    ft.polos_real = np.real(ft.polos)
    ft.sys_estable = es_estable(ft.polos_real)
    resp_escalon = ctl.step_info(ft.sys)
    ft.sobreimpulso= resp_escalon["Overshoot"]
    ft.p_time = resp_escalon["PeakTime"]
    print("Estabilidad del mejor = {}".format(ft.sys_estable))
    print("Sobreimpulso del mejor: {}".format(ft.sobreimpulso))
    print("T al sobreimpulso del mejor: {}".format(ft.p_time))
```

```python
###Se definen los parámetros del algoritmo genético

initial_individual = [2, 2.5, 3, 2, 3, 2, 2.5, 1.5, 1.5, 1.7,
 1, 2, 1, 1.5, 1]
fitness_function = fitness_func
num_generations = 20 #generaciones
num_parents_mating = 8 #número de padres
sol_per_pop = 50 #tamaño de población
num_genes = 15 #numero de genes
keep_parents = 5 #padres que se mantienen
parent_selection_type = "tournament" #Tipo de selección
crossover_type = "single_point" #Tipo de crossover
mutation_type = "random" #tipo de mutación
mutation_percent_genes = 50 #Porcentaje de mutaciones
initial_population = [] #Individuo inicial
for i in range(sol_per_pop):
    initial_population.append(initial_individual) 
#Espacio de alelos
space = []
for i in range(200):
    space.append(i/100+1)
gene_space = np.array(space)
```

```python
#Se crea una instancia del algoritmo genético dentro del modulo ga
ga_instance = pygad.GA(initial_population=initial_population,
                       num_generations=num_generations,
                       num_parents_mating=num_parents_mating, 
                       fitness_func=fitness_function,
                       sol_per_pop=sol_per_pop,
                       num_genes=num_genes,
                       parent_selection_type=parent_selection_type,
                       keep_parents=keep_parents,
                       crossover_type=crossover_type,
                       mutation_type=mutation_type,
                       mutation_percent_genes=mutation_percent_genes,
                       gene_space=gene_space,
                       on_generation=callback_generation)
```

```python
#Se corre la instancia del algoritmo genetico
ga_instance.run()
```

```python
#Se muestran los resultados obtenidos
ga_instance.plot_result()
# Returning the details of the best solution.
solution, solution_fitness, solution_idx = ga_instance.best_solution()
print("Parameters of the best solution : {solution}".format(solution=solution))
print("Fitness value of the best solution = {solution_fitness}".format(solution_fitness=solution_fitness))
print("Index of the best solution : {solution_idx}".format(solution_idx=solution_idx))
if ga_instance.best_solution_generation != -1:
    print("Best fitness value reached after {best_solution_generation} generations.".format(best_solution_generation=
                                                                                ga_instance.best_solution_generation))

mejor_sol=sacar_modelo(solution)
plt.figure(1)
T,yout= ctl.step_response(mejor_sol,60)

plt.xlabel("Tiempo")
plt.ylabel("Magnitud")
plt.plot(T.T, yout.T)
plt.show(block=False)
```
